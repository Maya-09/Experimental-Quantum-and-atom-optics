```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
```

```python
img = cv2.imread("Emoji.png", 0)
plt.imshow(img)
#gray scale image
```

```python
img = cv2.resize(img, (64, 64))
plt.imshow(img)
n = 64*64
#Resized image
```

```python
pattern = np.zeros((64, 64))
result = np.zeros((64, 64))
bn_seq = []
for i in range(64):
    for j in range(64):
        pattern[i][j] = 1  #set the jth pixel in the ith row to 1 indicating illumated pixel
        p1 = img* pattern  #object multiplication
        bn= p1.sum()       #bucket intensity value
        bn_seq.append(bn)
        result += bn*pattern   #image reconstruction
        pattern[i][j] = 0       #reset illuminated pixel
plt.imshow(result)
plt.title("Raster Emoji Image")  
    
   
    
```

```python
plt.imsave("raster_emoji.jpeg", result)
```

```python
meas = [i for i in range(n)]
```

```python
plt.plot(meas, bn_seq, "r")
plt.title("Intensity Distibution") 
```
