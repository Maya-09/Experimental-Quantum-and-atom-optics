```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab


```

```python
img = cv2.imread("flower2.jfif", 0)
```

```python
plt.imshow(img)
```

```python
img = cv2.resize(img, (64, 64))
n = 64*64
plt.imshow(img)
```

```python
patterns = np.random.randint(2, size =(n, n))
```

```python
result = np.zeros((64, 64))
int_seq = []
for i in range(n):
    p1 = patterns[i, :].reshape(64, 64)
    p2 =  1 - p1
    p = img*p1
    bn1 = p.sum()
    p3 = img*p2
    bn2 = p3.sum()
    int_fluc = bn1 - bn2
    int_seq.append(int_fluc)
    result += int_fluc*p1
plt.imshow(result)
plt.title("Random Flower Image")  

   
    
```

```python
plt.imsave("random_flower.jpeg", result)
```

```python
meas = [i for i in range(n)]
```

```python
plt.plot(meas, int_seq, "r")
plt.title("Intensity Fluctuation Distribution")
```
