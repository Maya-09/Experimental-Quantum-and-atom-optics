## EXERCISE 1

Exercise 1a

```python
from matrix_functions import matrix_lens, fresnel_prop_matrix, phase_quad_matrix 
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
f1 = 1
x = np.linspace(-1, 1, N)
H = matrix_lens(f1)
def input_field():
    Ein = np.zeros((N, 1))
    for i in range(350, 651):
        Ein[i] = 1
    return Ein
E_in = input_field()
E_out = np.matmul(H, E_in)
I_out = np.abs(E_out)**2
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, I_out,"r")
ax.set_xlabel("x")
ax.set_ylabel("Intensity")
ax.grid(True)
ax.xaxis.set_ticks(np.arange(-1, 1, 0.2))
plt.show
```
